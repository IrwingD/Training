Red line is the decision boundary seperating the classes '+' and '-'

W is the vector pointing perpendicular to the seperating hyperplane (decision boundary)
U is the vector to the unknown new point U


The way SVM works is by projecting vector U onto W and figures out on which side of the hyperplane vector U is

it does this by doing a dot product of vectors U and W

### The Logic

The vector U is made of features X1 and X2 we can denote this as :

$X_{-SV}$ -> features for all the '-' points and

$X_{+SV}$ -> features for all the '+' points


##### Now 
$\vec{U}$.$\vec{W}$+b >=0 for all '+' ,therefore

$\vec{X_{+SV}}$.$\vec{W}$+b=+1


##### AND 

$\vec{U}$.$\vec{W}$+b <=0 for all '-' ,therefore

$\vec{X_{-SV}}$.$\vec{W}$+b=-1


We can thus conclude that $y_{i}$ will be either +1 or -1 depending on if the unknown vector is of class'+' or '-'
